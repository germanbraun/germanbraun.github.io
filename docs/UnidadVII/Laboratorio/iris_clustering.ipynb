```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Create a DataFrame
df = pd.DataFrame(X, columns=feature_names)
df.columns = ['longitud del sépalo', 'ancho del sépalo', 'longitud del pétalo', 'ancho del pétalo']

# Apply K-means with 3 clusters, as the dataset has 3 species
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(X)
df['Cluster'] = kmeans.labels_

# Create a scatter plot using 'longitud del sépalo' and 'longitud del pétalo' for visualization
plt.figure(figsize=(10, 7))
plt.scatter(df['longitud del sépalo'], df['longitud del pétalo'], c=df['Cluster'], s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 2], c='red', s=200, alpha=0.7, label='Centroides')

plt.title('Clustering K-Means en el Conjunto de Datos Iris')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Longitud del Pétalo (cm)')
plt.legend()
plt.savefig('iris_kmeans_clustering.png')

print("Plot guardado como 'iris_kmeans_clustering.png'")
```
