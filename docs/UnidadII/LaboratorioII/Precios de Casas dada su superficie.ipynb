# Paquetes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Generación de datos sintéticos

```python
# Datos sintéticos
np.random.seed(42)
X = np.random.randint(50, 300, size=50).reshape(-1, 1)  # superficie en m²
y = 1500 * X.flatten() + np.random.normal(0, 20000, size=50)  # precio con ruido
```

# Dataset generado

```python
# Crear DataFrame para mostrar
df = pd.DataFrame({
    "Superficie (m²)": X.flatten(),
    "Precio (USD)": y.astype(int)  # enteros para mostrar
})
print("\n=== Dataset generado ===")
print(df)
```

# Entrenamiento

```python
# Entrenar modelo
model = LinearRegression()
model.fit(X, y)
```

# Visualización de Resultados

```python
# Predicciones para graficar
X_pred = np.linspace(50, 300, 100).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Resultados del modelo
print(f"\nPendiente (USD/m²): {model.coef_[0]:.2f}")
print(f"Intercepto: {model.intercept_:.2f}")

plt.scatter(X, y, color="blue", label="Datos reales")
plt.plot(X_pred, y_pred, color="red", label="Regresión lineal")
plt.xlabel("Superficie (m²)")
plt.ylabel("Precio (USD)")
plt.title("Predicción de precio según superficie")
plt.legend()
plt.show()
```

# Predicción

```python
# Superficie de test
superficie_test = np.array([[180]])  # doble corchete porque es 2D

# Predicción
precio_estimado = model.predict(superficie_test)

print(f"Precio estimado para 180 m²: {precio_estimado[0]:.2f} USD")
```
