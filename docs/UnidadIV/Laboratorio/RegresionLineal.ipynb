```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --------------------------
# 1. Generar datos sintéticos
# --------------------------
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)           # X entre 0 y 2
true_theta0 = 1
true_theta1 = 3
y = true_theta0 + true_theta1 * X + np.random.randn(m, 1)  # ruido

# --------------------------
# 2. Crear y entrenar modelo
# LinearRegression() ya computan los valores para theta (sin usar el método del gradiente)
# --------------------------
model = LinearRegression()
model.fit(X, y)

# Coeficientes aprendidos
theta0 = model.intercept_[0]
theta1 = model.coef_[0,0]

print(f"Theta0 (intercept): {theta0:.3f}")
print(f"Theta1 (pendiente): {theta1:.3f}")

# --------------------------
# 3. Predicción
# --------------------------
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

# --------------------------
# 4. Visualización
# --------------------------
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Recta ajustada')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal con scikit-learn')
plt.legend()
plt.show()
```
