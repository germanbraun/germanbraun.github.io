```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

# Dataset sintético
np.random.seed(0)
n = 50
X1 = np.random.rand(n, 1) * 10
X2 = np.random.rand(n, 1) * 10
y = 3*X1.squeeze() + 0.5*X2.squeeze() + np.random.randn(n)*2
X = np.hstack([X1, X2])

# Distintos valores de lambda
lambdas = np.logspace(-3, 1, 40)
coefs = []

for l in lambdas:
    lasso = Lasso(alpha=l, max_iter=10000)
    lasso.fit(X, y)
    coefs.append(lasso.coef_)

coefs = np.array(coefs)

# Gráfico
plt.figure(figsize=(7,5))
plt.plot(lambdas, coefs[:,0], label="w1 (x1)")
plt.plot(lambdas, coefs[:,1], label="w2 (x2)")
plt.xscale("log")
plt.xlabel("λ (alpha en sklearn)")
plt.ylabel("Coeficientes")
plt.title("Trayectoria de coeficientes con LASSO")
plt.legend()
plt.grid(True)
plt.show()
```
