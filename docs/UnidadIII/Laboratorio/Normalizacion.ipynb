```python
import matplotlib.pyplot as plt
import numpy as np

# Datos originales
x = np.array([5, 7, 3, 7, 10])
indices = np.arange(len(x))

# Calcular media y desviación estándar muestral
mean = x.mean()
std = x.std(ddof=1)

# Normalizar los datos (z-score)
z = (x - mean) / std

# Crear gráfico
plt.figure(figsize=(10, 5))

# Gráfico de datos originales
plt.subplot(1, 2, 1)
plt.stem(indices, x)
plt.axhline(mean, color='red', linestyle='--', label=f'Media = {mean:.2f}')
plt.title("Datos Originales")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)

# Gráfico de datos normalizados
plt.subplot(1, 2, 2)
plt.stem(indices, z)
plt.axhline(0, color='red', linestyle='--', label='Media = 0')
plt.title("Datos Normalizados")
plt.xlabel("Índice")
plt.ylabel("Valor Normalizado (Z-score)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
